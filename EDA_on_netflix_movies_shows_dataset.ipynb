```python
#importing libraries for EDA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/netflix_titles.csv.zip")
df.head()
```

```python
df.tail(15)
```

```python
df.dtypes
```

```python
pd.unique(df['date_added'])
```

```python
df=df.drop(['date_added'],axis=1)
df.head()
```

```python
df=df.rename(columns={"listed_in":"genre","show_id":"id"})
df.head()
```

```python
df.shape
```

```python
duplicate_rows_df=df[df.duplicated()]
print("Number of duplicate rows : ",duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df=df.drop_duplicates()
df.head()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df=df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
sns.boxplot(x=df['release_year'])
```

```python
df.release_year.value_counts().nlargest(20).plot(kind='bar',figsize=(10,5))
plt.title("Number of Movies or TV shows by release year")
plt.ylabel("Number of Movies or TV shows")
plt.xlabel("Release year")
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['release_year'])
plt.xlabel("Release year")
plt.ylabel("Number of movies or TV shows")
```

```python
plt.figure(figsize=(5,5))
sns.scatterplot(x=df['type'],y=df['release_year'])
```

```python
plt.figure(figsize=(5,5))
sns.lineplot(x=df['type'],y=df['release_year'])
```
